```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sub_data =  data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
corr = sub_data.corr()
fig, ax = plt.subplots(figsize=(10, 8))
colormap = sns.diverging_palette(220, 10, as_cmap = True)
dropvals = np.zeros_like(corr)
dropvals[np.triu_indices_from(dropvals)] = True
sns.heatmap(corr, cmap = colormap, linewidths = .5, annot = True, fmt = ".2f", mask = dropvals)
plt.show()
```

Body mass has a strong correlation with flipper length.
